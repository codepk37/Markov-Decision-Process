```python


import copy

import numpy as np
#bottom left is origin
#so at accessing subtracted 3-y , to access correct index
def  Utility(action ,ori_x,ori_y, M ,p):
    r= len(M)
    c=len(M[0])
    if M[ori_x][ori_y]==-1: #Penalty 
        return -1
    if M[ori_x][ori_y]==1: #Reward ,can be skipped in loop, wrote for safely
        return 1
    if M[ori_x][ori_y]=='WALL':
        return 'WALL'
    
    if(action=="UP"):
        move=(1,0)
        tot=-0.04 #step cost 
        if(ori_y-1<0  or M[ori_x][ori_y-1]=='WALL'):
            tot+=  (0.95)*(1-p)*(1/2)*M[ori_x][ori_y]
        elif(ori_y-1>=0):
            tot+=  (0.95)*(1-p)*(1/2)*M[ori_x][ori_y-1]

        if( ori_y+1>=c or M[ori_x][ori_y+1]=='WALL'):
            tot+=  (0.95)*(1-p)*(1/2)*M[ori_x][ori_y]
        elif( ori_y+1<c):
            tot+= (0.95)*(1-p)*(1/2)*M[ori_x][ori_y+1]

        if( ori_x+1 >=r or M[ori_x+1][ori_y]=='WALL'):
            tot+=  (0.95)*(p)*M[ori_x][ori_y]
        elif( ori_x+1 <r):
            # print("    --------   ",M[ori_x][ori_y] )
            tot+=  (0.95)*(p)*M[ori_x+1][ori_y]
        return tot
        
    elif(action=="DOWN"):
        move=(-1,0)
       
        tot=-0.04
        if(ori_y-1<0 or M[ori_x][ori_y-1]=='WALL'):
            tot+=  (0.95)*(1-p)*(1/2)*M[ori_x][ori_y]
        elif(ori_y-1>=0):
            tot+= (0.95)*(1-p)*(1/2)*M[ori_x][ori_y-1]
        if( ori_y+1>=c or M[ori_x][ori_y+1]=='WALL'):
            tot+=  (0.95)*(1-p)*(1/2)*M[ori_x][ori_y]
        elif( ori_y+1<c):
            tot+=  (0.95)*(1-p)*(1/2)*M[ori_x][ori_y+1]
        if( ori_x-1 <0 or M[ori_x-1][ori_y]=='WALL'):
            tot+=  (0.95)*(p)*M[ori_x][ori_y]
        elif( ori_x-1 >=0):
            tot+=  (0.95)*(p)*M[ori_x-1][ori_y]
        return tot


    
    elif(action=="LEFT"):
        move=(0,-1)
       
        tot=-0.04
        if(ori_x-1<0 or M[ori_x-1][ori_y]=='WALL'):
            tot+=  (0.95)*(1-p)*(1/2)*M[ori_x][ori_y]
        elif(ori_x-1>=0):
            tot+=  (0.95)*(1-p)*(1/2)*M[ori_x-1][ori_y]
        if( ori_x+1>=r or  M[ori_x+1][ori_y]=='WALL'):
            tot+=  (0.95)*(1-p)*(1/2)*M[ori_x][ori_y]
        elif( ori_x+1<r):
            tot+=  (0.95)*(1-p)*(1/2)*M[ori_x+1][ori_y]
        
        if( ori_y-1 <0 or M[ori_x][ori_y-1]=='WALL'):
            tot+=  (0.95)*(p)*M[ori_x][ori_y]
        elif( ori_y-1 >=0):
            tot+=  (0.95)*(p)*M[ori_x][ori_y-1]
        return tot
        

    elif(action=="RIGHT"):
        move=(0,1)
     
        tot=-0.04
        if(ori_x-1<0 or M[ori_x-1][ori_y]=='WALL'):
            tot+=  (0.95)*(1-p)*(1/2)*M[ori_x][ori_y]
        elif(ori_x-1>=0):
            tot+=  (0.95)*(1-p)*(1/2)*M[ori_x-1][ori_y]
        if( ori_x+1>=r or M[ori_x+1][ori_y]=='WALL'):
            tot+=  (0.95)*(1-p)*(1/2)*M[ori_x][ori_y]
        elif( ori_x+1<r):
            tot+=  (0.95)*(1-p)*(1/2)*M[ori_x+1][ori_y]
        
        if( ori_y+1 >=c or M[ori_x][ori_y+1]=='WALL'):
            tot+=  (0.95)*(p)*M[ori_x][ori_y]
        elif( ori_y+1 <c):
            tot+=  (0.95)*(p)*M[ori_x][ori_y+1]
        return tot



def value_iterations(M,Actions,p):
    lasV= M.copy()
    forma= copy.deepcopy(M)
    cnt=0
    while True :#and cnt<1000:
        cnt+=1
        oldV= lasV.copy()
        delta = 0
        for r in range(4):
            for c in range(3):
                if(oldV[r][c]=='WALL'):
                    continue

                a1=Utility(Actions[0],r,c,oldV,p)
                a2=Utility(Actions[1],r,c,oldV,p)
                a3=Utility(Actions[2],r,c,oldV,p)
                a4=Utility(Actions[3],r,c,oldV,p)

                maxi=new_value = max(a1, a2, a3, a4)
                delta = max(delta, abs(oldV[r][c] - new_value))
                lasV[r][c] = new_value

                ####
                if(M[r][c]=='WALL' or M[r][c]==-1 or M[r][c]==1):
                    forma[r][c]= 'none'
                elif(maxi==a1):
                    forma[r][c]='down'
                elif(maxi==a2):
                    forma[r][c]= 'up'
                elif(maxi==a3):
                    forma[r][c]= 'left'
                elif(maxi==a4):
                    forma[r][c]='right'
                

        # print(lasV)
        if delta < 0.000001 :
            break
        

    return lasV,forma


M = [
    [0,-1,1],
    [0,0,0],
    [0,'WALL',0],
    [0,0,0]
]

Actions=["UP","DOWN","LEFT","RIGHT"]



```

```python
#TASK A
print("TASK A\n")
p=0.7
z,f=value_iterations(M,Actions,p)
for row in z:
    print(row)
```

```python
#TASK B
print("TASK B")

import copy

def main():
    for p in range(1, 10): 
        p = p / 10.0
        print(f"\np={p}")
        z,forma=value_iterations(M,Actions,p)
        for row in z:
            print(row)
        
        for f in forma:
            print(f)
        
        
        

main()




```



### Observing the policies for different values of \( p \), we can notice several trends and changes:

1. **Direction of Movement:**
   - As \( p \) increases, there is a tendency for the optimal policy to favor movement towards the goal state.
   - For lower values of \( p \), the agent is more likely to explore various directions, including movements perpendicular to the intended action.

2. **Avoidance of Red State:**
   - The policy consistently avoids the red state as it incurs a penalty of -1. This is evident across all \( p \) values.

3. **Efficiency vs. Exploration:**
   - For smaller \( p \) values (e.g., 0.1 to 0.4), the policies exhibit more exploratory behavior, often choosing actions that might not lead directly towards the goal but allow for better exploration of the environment.
   - As \( p \) increases (e.g., from 0.5 to 0.9), the policies tend to become more deterministic, focusing on reaching the goal state efficiently.

4. **Adaptation to Probabilistic Nature:**
   - With higher \( p \) values, the policy tends to rely more on the probabilities associated with actions. This is reflected in the directional choices, favoring actions with higher probabilities of success.
   - Lower \( p \) values lead to more varied policies, as the agent has less certainty about the outcome of its actions due to the increased likelihood of moving in directions perpendicular to the intended action.

5. **Impact of Discount Factor \( \gamma \) (Gamma):**
   - The discount factor \( \gamma \) influences the balance between immediate rewards and future rewards. In this case, \( \gamma = 0.95 \) encourages the agent to prioritize long-term rewards.
   - The policies reflect this by favoring actions that maximize expected cumulative rewards over time, especially as \( p \) increases.
